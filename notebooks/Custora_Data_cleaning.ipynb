```python
#Load Dataset
df = pd.read_csv("/content/drive/MyDrive/Custora-RFM-Based Customer Segmentation & Retention Analysis/online_retail.csv", encoding="ISO-8859-1")
df.head()
```

```python
#Basic Understanding (Interview Gold ðŸ¥‡)
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
#Data Cleaning
# Remove missing CustomerID
df = df.dropna(subset=["CustomerID"])

# Remove cancelled orders
df = df[~df["InvoiceNo"].str.startswith("C")]

# Remove negative or zero quantity
df = df[df["Quantity"] > 0]

# Remove zero or negative price
df = df[df["UnitPrice"] > 0]
```

```python
#Convert Date
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```
